    Based on 30 SMA and 100 SMA, buy sell signal generated. 

    Reliance 20 year data has been taken for analysis

```python
import pandas as pd
import numpy as np
import datetime as datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# load reliance data
path = '/home/debashis/Data/Finance_Data/'
df = pd.read_csv(path+'RELIANCE_20yr_data_nse.csv', low_memory=False)
df.head()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.head()
```

```python
# create SMA 30
sma30 = pd.DataFrame()
sma30["Adj_Close_Price"] = df["Adj Close"].rolling(window = 30).mean()
sma30
```

```python
ema30 = pd.DataFrame()
ema30["Adj_Close_Price"] = df.ewm(com = 0.5, min_periods = 30).mean()['Adj Close']
ema30
```

```python
ema100 = pd.DataFrame()
ema100["Adj_Close_Price"] = df.ewm(com = 0.5, min_periods = 100).mean()['Adj Close']
ema100
```

```python
# create sma 100
sma100 = pd.DataFrame()
sma100["Adj_Close_Price"] = df["Adj Close"].rolling(window = 100).mean()
sma100
```

```python
# Visualise data
plt.figure(figsize = (16, 6))
plt.plot(df["Adj Close"], label = 'Reliance Close Price')
plt.plot(sma30["Adj_Close_Price"], label = 'SMA 30')
plt.plot(sma100["Adj_Close_Price"], label = 'SMA 100')
plt.title('Reliance Close Price, SMA 30 and SMA 100')
plt.xlabel('17th May 2000 to 15th May 2020')
plt.ylabel('Adj Close Price in Rupees')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# Create a new dataframe to store adj close price and sma30 and sma100
new_data = pd.DataFrame()
new_data['Adj_Close_Price'] = df['Adj Close']
new_data['SMA30'] = sma30.Adj_Close_Price
new_data['SMA100'] = sma100.Adj_Close_Price
new_data.head()
```

```python
# function which give signal for buy and sell
def buy_sell(data):
    sig_buy = []
    sig_sell = []
    flag = -1
    
    for i in range(len(data)):
        if data['SMA30'][i] > data['SMA100'][i]:
            if flag != 1:
                sig_buy.append(data['Adj_Close_Price'][i])
                sig_sell.append(np.nan)
                flag = 1
            else:
                sig_buy.append(np.nan)
                sig_sell.append(np.nan)
                
        elif data['SMA30'][i] < data['SMA100'][i]:
            if flag != 0:
                sig_buy.append(np.nan)
                sig_sell.append(data['Adj_Close_Price'][i])
                flag = 0
            else:
                sig_buy.append(np.nan)
                sig_sell.append(np.nan)
        else:
            sig_buy.append(np.nan)
            sig_sell.append(np.nan)
            
    return (sig_buy, sig_sell)
```

```python
# Now collect buy sell price
buy_sell_data = buy_sell(new_data)
new_data['Buy_Signal_Price'] = buy_sell_data[0]
new_data['Sell_Signal_Price'] = buy_sell_data[1]
```

```python
plt.figure(figsize = (16, 6))
plt.plot(df["Adj Close"], label = 'Reliance Close Price', alpha = 0.35)
plt.plot(sma30["Adj_Close_Price"], label = 'SMA 30', alpha = 0.35)
plt.plot(sma100["Adj_Close_Price"], label = 'SMA 100', alpha = 0.35)
plt.scatter(new_data.index, new_data['Buy_Signal_Price'], label = 'Buy', marker='^', color = 'green')
plt.scatter(new_data.index, new_data['Sell_Signal_Price'], label = 'Sell', marker = 'v', color = 'red')
plt.title('Reliance Close Price, SMA 30 and SMA 100')
plt.xlabel('17th May 2000 to 15th May 2020')
plt.ylabel('Adj Close Price in Rupees')
plt.legend(loc = 'upper left')
plt.show()
```
